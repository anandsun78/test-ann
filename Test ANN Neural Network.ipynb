```python
# Part 1: Data Preprocessing 
# STEP 1: Importing the libraries
import numpy as np
import pandas as pd


# STEP 2: Importing the datasets
dataset = pd.read_excel('TrainANN.xlsx')

X = dataset.iloc[:, 0:6].values
# X:  0:day,  1:in-terminal,  2:recommended charging times,  3:hours needed to charge
y = dataset.iloc[:, 6:9].values
# y: theoratical charging schedule

# STEP 3: Encode categorial data (Not Needed)

# STEP 4: Splitting dataset into the training set and test set (Automatically)
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.12, random_state = 0)
# Test set: 12%


#Step 5: Feature Scaling (Puts all the variables on the same scale ranging b/w -1/1)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train_scale = sc.fit_transform(X_train[:, 0:3])
X_train[:, 0:3] = X_train_scale

X_test_scale = sc.transform(X_test[:, 0:3])
X_test[:, 0:3] = X_test_scale


# Part 2: Making the ANN
#Importing the Keras library and required packages
import keras
from keras.models import Sequential # Seq module: Required to initialize our Neural Network
from keras.layers import Dense #Dense module: required to build the layers of our ANN

# Initializing the ANN by defining the ANN as a sequence of layers 
classifier = Sequential()

# Adding the input layer and the first hidden layer 
classifier.add(Dense(output_dim = 4, init = 'uniform', activation = 'relu', input_dim = 6))  # classifier.add(layers) "add method" used to add the different layers in our NN 
# First Hidden Layer Nodes: 5
# Initializer: uniform
# Activation function: relu
# Input dimensions: 6

# Adding the output layer 
classifier.add(Dense(output_dim = 3, init = 'uniform', activation = 'sigmoid'))
# Output dimensions: 3
# Initializer: uniform
# Activation function: sigmoid

# Compiling the ANN (by applying stochastic gradient descent on the whole ANN)
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
# Optimizer: adam
# Loss: binary_crossentropy
# metrics: accuracy (can alose use: mean squared error, mean absolute error, mean absolute percentage error)

# Fitting the ANN to our training set
classifier.fit(X_train, y_train, batch_size = 24, epochs = 100) #See Step 6 and 7 in document
# Batch size = 24
# Epochs = 100


# Part 3: Evaluating the ANN
# Predicting the Test set results
y_pred = classifier.predict(X_test)

# Predicting single results
datasetPredict1 = pd.read_excel('PredictThisDay1.xlsx')
X_newpred1 = datasetPredict1.iloc[:, 1:7].values
new_prediction1 = classifier.predict(X_newpred1)


```
